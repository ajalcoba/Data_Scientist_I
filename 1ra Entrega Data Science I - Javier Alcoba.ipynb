### Nivel de Educación alcanzado de los empleados que dejaron la empresa vs los que no la dejaron:

```python

# Calcular los porcentajes de renuncia y no renuncia por nivel de educación

df_no_renuncia_conteo = df[df['Resigned'] == False]['Education_Level'].value_counts(normalize=True)
df_renuncia_conteo = df_renuncia['Education_Level'].value_counts(normalize=True)

# Crear un subplot con dos gráficos de torta
fig, axes = plt.subplots(1, 2, figsize=(10, 5))

# Gráfico de torta para empleados que no renunciaron
axes[0].pie(df_no_renuncia_conteo, labels=df_no_renuncia_conteo.index,autopct='%1.1f%%', startangle=90)
axes[0].set_title('No renunciaron')

# Gráfico de torta para empleados que renunciaron
axes[1].pie(df_renuncia_conteo, labels=df_renuncia_conteo.index,autopct='%1.1f%%', startangle=90)
axes[1].set_title('Renunciaron')

plt.suptitle('Renuncias vs Educación alcanzada', fontsize=16)
plt.show()
```

Como se puede observar, no existe prácticamente diferencia alguna del nivel de educación y si el empleado dejó o no la empresa.

### Empleados que renunciaron vs género:

```python
# Agrupación de empleados por género

df_no_renuncio_genero = df[df['Resigned'] == False].groupby('Gender').size()
df_renuncio_genero = df[df['Resigned'] == True].groupby('Gender').size()


# Gráficos de barra
plt.figure(figsize=(10, 6))
df_no_renuncio_genero.plot(kind='bar', rot=0, label='No renunció', color = 'blue')
df_renuncio_genero.plot(kind='bar', rot=0, label='Renunció', color='red')

plt.xlabel('Género')
plt.ylabel('Cantidad')
plt.title('Comparación de renuncias por género')
plt.legend()
plt.show()
```

En este caso en particular, se puede observar que el nivel de hombres que dejaron la empresa fueron, en proporción, idénticos al nivel de mujeres que dejaron la empresa. Lo mismo se da para el género "Otros". La proporción de cada género fue de, aproximadamente, 10 a 1 respecto a quedarse en la empresa respecto a irse de la empresa.

```python
# Filtrar los datos para sueldos en función del género 
df_hombre = df[df['Gender'] == 'Male']
df_mujer = df[df['Gender'] == 'Female']
df_otro = df[df['Gender'] == 'Other']

# Subplot con dos boxplots dentro en forma paralela
fig, axes = plt.subplots(3, 1, figsize=(12, 18))

# Boxplot para Hombres
axes[0].boxplot(df_hombre['Monthly_Salary'], vert=False)
axes[0].set_title('Sueldo para empleados hombres')
axes[0].set_xlabel('Sueldo para empleados hombres')

# Boxplot para Mujer
axes[1].boxplot(df_mujer['Monthly_Salary'], vert=False)
axes[1].set_title('Sueldo para empleados mujeres')
axes[1].set_xlabel('Sueldo para empleados mujeres')

# Boxplot para Otro tipo
axes[2].boxplot(df_otro['Monthly_Salary'], vert=False)
axes[2].set_title('Sueldo para empleados de otro género')
axes[2].set_xlabel('Sueldo para empleadosde otro género')

plt.show()
```

Como se puede ver, al igual que para gráficos anteriores, el salario se mantiene invariable respecto a si es de un hombre, mujer u otro género.

```python
# Matriz de correlación de variables
def plotCorrelationMatrix(df, graphWidth):
    nombre = 'Dataset'
    corr = df.corr()
    plt.figure(num=None, figsize=(graphWidth, graphWidth), dpi=80, facecolor='w', edgecolor='k')
    corrMat = plt.matshow(corr, fignum = 1)
    
    plt.xticks(range(len(corr.columns)), corr.columns, rotation=90)
    plt.yticks(range(len(corr.columns)), corr.columns)
    
    plt.gca().xaxis.tick_bottom()
    plt.colorbar(corrMat)
    
    plt.title(f'Matriz de correlaciones del {nombre}', fontsize=15)
    plt.show()
```

```python
plotCorrelationMatrix(df,10)
```

Como se puede ver en la matriz de correlaciones entre variables, es que este dataset no sirve prácticamente para nada. No hay casi ninguna distinción entre variales ni parece que alguna de ellas pueda tener que ver con la migración de empleados en la empresa.
